```python
import random
import string
import datetime
import csv

def generate_data_and_write_csv(num_rows, filename):
    with open(filename, mode='w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['order_id', 'food_id', 'booking_time', 'id', 'quantity'])
        for i in range(num_rows):
            order_id = "FFB" + str(random.randint(1, 500)).zfill(5)
            food_id = "food" + str(random.randint(1, 45)).zfill(2)
            booking_time = datetime.datetime.now() - datetime.timedelta(days=random.randint(0, 30), hours=random.randint(0, 23), minutes=random.randint(0, 59), seconds=random.randint(0, 59))
            user_id = "ID" + str(random.randint(1, 100)).zfill(4)
            quantity = random.randint(1, 5)
            row = [order_id, food_id, booking_time, user_id, quantity]
            writer.writerow(row)
```

```python
generate_data_and_write_csv(10000, 'data.csv')
```
